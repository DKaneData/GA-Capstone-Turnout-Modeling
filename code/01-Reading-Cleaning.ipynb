```python
# A vote history function to binarize vote history where if vote history is missing / 0, it returns 0. But if there's any vote history present (voted, voted absentee, voted in the primary), it returns a 1.
def vh_recode(number):
    if number == 0:
        number = 0
    else:
        number = 1

    return number
```

```python
vh_cols
```

```python
# Returns last letter
'vh23g'[-1]
```

```python
# Returns 2nd to last 2 letters
'vh23g'[-3:-1]
```

```python
# combines last letter & 2nd to last 2 letters
'vh23g'[-1] + 'vh23g'[-3:-1]
```

```python
vh_recode_df = []
```

```python
#For each column in the vote history columns
for col in vh_cols.columns: 
    
    # New_col_name is the last letter of the column plus the 2 letters before that. So vh23g becomes g23 ... vh10p becomes p10
    new_col_name = col[-1] + col[-3:-1]

    #Apply the recoding function to that column and output it onto vh_cols with the new column name
    vh_cols[new_col_name] = vh_cols[col].apply(vh_recode)
```

```python
pd.set_option('display.max_columns', 60)
vh_cols.sample(7)
```

```python
# Get the last 28 columns of vote history that are the recoded vote history columns - from g23 to p10
recoded_vh = vh_cols[vh_cols.columns[-28:]]
recoded_vh
```

```python
# Joining the recoded vote history onto the main voter file
vf = vf.join(recoded_vh)
```
